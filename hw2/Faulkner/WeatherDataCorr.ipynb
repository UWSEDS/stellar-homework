## Bringing back the Fremont Bridge bicycle data

(1)  Loading the bridge data into a dataframe while parsing dates, then setting the date as the index:

```python
bdata = pd.read_csv('~/Documents/CSE599/HWlocal/HW2/Data/4xy5-26gy.csv',parse_dates=['date'])
bdf = pd.DataFrame(bdata)
bdf2 = bdf.set_index('date')
bdf2.index.name=None
bdf2
```

(2) Resampling the date index to show totals for each day (I'm also totaling nb and sb):

```python
bdf3 = bdf2.resample('1D',how='sum')
bdf3['total_riders'] = bdf3['fremont_bridge_nb'] + bdf3['fremont_bridge_sb']
bdf3
```

(3) Plotting total riders as a function of the date:

```python
bdf3['total_riders'].plot()
```

The above graph looks terrible because of the big gap in available data in our date range.  If we just want to see what the total rider profile is for the 2015 data, we can do the following, much like we did in HW1:

```python
b2data = pd.read_csv('~/Documents/CSE599/HWlocal/HW2/Data/4xy5-26gy.csv',parse_dates=['date'])
b2data['total_riders'] = b2data['fremont_bridge_nb'] + b2data['fremont_bridge_sb']
times = pd.DatetimeIndex(b2data['date'])
b2data['year']=times.year
b2df = pd.DataFrame(b2data)
b2df2 = b2df.set_index('date')
b2df2.index.name=None
b2df3 = b2df2[b2df2.year == 2015]
b2df4 = b2df3.resample('1D',how='sum')
b2df4['total_riders'].plot()
```

Referencing a calendar, it looks like traffic peaks during the weekdays, so it's probable that most traffic is commuting.

## Joining the two datasets

(1) Figuring out the join() command:

```python
fdf = wdf2.join(bdf3)
fdf
```

(2) Scatter plots of:

high temperature in tenths of degree C,

low temperature in tenths of degree C,

precipitation in tenths of mm,

average wind speed in tenths of m/s,

maximum 2-minute wind speed in tenths of m/s,

snowfall in mm, and

direction of the fastest 2-minute wind:

```python
fdf.plot.scatter('TMAX','total_riders')
fdf.plot.scatter('TMIN','total_riders')
fdf.plot.scatter('PRCP','total_riders')
fdf.plot.scatter('AWND','total_riders')
fdf.plot.scatter('WSF2','total_riders')
fdf.plot.scatter('SNOW','total_riders')
fdf.plot.scatter('WDF2','total_riders')
```

It appears that temperature strongly influences whether people go out riding on the Fremont Bridge, with the biggest riding days falling between high temps of 20-35 C and low temps of 12-18 C.  This is a bit of a fallacy, though, since we have seen that our 2015 data are limited to July and August, while our much smaller 2012 data are from October through December.  Some of this difference could be due to the temperature, but some of it could also be due to an increase in cycling popularity (or infrared counter technology, perhaps?  I don't know if it was replaced during the data gap) between 2012 and 2015.  Also note that there are many warm days with low numbers of riders: as we saw above, these are likely weekends in Jul/Aug 2015.

Similarly, the days with the largest numbers of riders correlate with days of zero precipitation.  This is tied to the big numbers in the 2015 data, since the months available for that year (Jul/Aug) had very low precipitation, as seen in the first graph of this notebook.  So most of the data points that have some precipitation are from the 2012 rider data.

The correlations with wind speed are likely more reliable than those with precipitation or temperature.  Scatter plots 4 and 5 show somewhat of the same trend: there are no days with large numbers of riders where the wind averages over 4.5 m/s (10 mph) or 2-minute-maxes over 8.5 m/s (19 mph).  Not many people like to ride in strong winds, apparently.  Note also that there are plenty of days with low rider totals and low winds: these are likely weekends from both 2012 and 2015.  It would be interesting to look at only weekdays during this analysis (or at least separate it into weekends and weekdays).  Maybe that'll be HW3.

In scatter plot 6, I tried to show that no one wants to ride in snow, but there were no snowy days in the dates included in the rider data.

Scatter plot 7 shows almost no correlation between number of riders and wind direction.  I guess we shouldn't be surprised that riders heading for the Fremont Bridge don't care which way the wind is blowing at the airport.
