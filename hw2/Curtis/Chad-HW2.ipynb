# Scatter plot of total riders per day as a function of the the average daily wind speed (m/s)

# Scatter plot of the total riders per day as a function of the fastest 2-minute wind speed (m/s)

# Scatter plot of the total riders per day as a function of the direction of fastest 2-minute wind speed (degrees)

```python
combine.plot(kind="scatter", x="WDF2", y="total")
```

The most obvious trend in the data is related to precipitation.  Many bikers crossed the bridge on days with zero precipitation, with very few, if any, riders on days with precipitation.

The temperature data was less cut and dry, but with clear threshold values.  If the maximum temperature dropped below about 20C or if the minimum temperature dropped below 13C, very few bikers crossed the bridge.  There wasn't a similar cutoff at the highs in temperature.  Apparently it doesn't get hot enough in Washington to discourage bikers from getting out on the road.

Trends with the availaable wind data weren't as clear cut, but it can be observed that there tends to be fewer bikers out at higher wind speeds.  The direction of the wind has no observable effect on bikers; I would have been surprised if there was.



Many of the data fields recorded the weather, creating a binary system of data (e.g. either it was foggy or it wasn't foggy).  This data could be used to see if certain weather conditions affected the number of bikers on the road.  Here were the findings:

Fog had a weak negative correlation with the number of bikers.  Most bikers would brave the decreased visibility to get to work.

There were very few days recorded with thunder (2 data points), but on those days, there were still quite a few bikers, but in decreased numbers. 

Only one recorded day of hail, and ~500 bikers still braved the storm.

Generally, no bikers rode on days with snow.
