### Second Question
The second question that we are going to explore is how ride counts vary by gender and weather conditions. As we saw above, there are not many riders that identify as other, so it is difficult to discern any interesting trends. Therefore, we will only focus on male and female riders in this discussion relating to gender and weather. Interacting with the plot below, we see a similar trend for both male and female riders when looking at daily number of riders versus wind speed. The most important thing we are able to observe from this plot is that they are less riders of both genders when there are greater wind speeds. The plot also suggests that wind speed does not seem to have a large effect on the number of riders since rider counts vary greatly at any wind speed. For example, at 4 MPH, there was a day with approximately 350 male riders, but there was also a day with close to zero riders at this same wind speed. Next, we can explore how precipitation affects daily rider counts. As with wind speed, there are less riders when there is more precipitation. There is a definite increase in ride counts among males as the temperature increases. Interestingly, the temperature does not seem to effect female ride counts as drastically. For both male and female riders, there were definitely less riders when the temperature was colder. From looking at these different plots, we were able to conclude that precipitation and temperature seem to have the biggest impact on daily ride count.
